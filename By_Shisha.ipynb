```python
# Definir el rango de fechas
start_date = "2022-01-01"
end_date = pd.Timestamp.today().strftime("%Y-%m-%d")  # Fecha actual

# Generar 500 fechas aleatorias dentro del rango
fechas_random = pd.to_datetime(np.random.choice(pd.date_range(start_date, end_date), size=500))

# Convertir a DataFrame
df_ventas = pd.DataFrame({"Fecha_Venta": fechas_random})

# Ordenar las fechas de manera ascendente
df_ventas = df_ventas.sort_values(by="Fecha_Venta").reset_index(drop=True)

# Mostrar las primeras filas
print(df_ventas.head())
```

```python
# generamos datos con una distribución normal para la columna cantidad
df_ventas['cantidad'] = np.random.normal(loc=10, scale=2, size=len(df_ventas))
# columna cantidad debe ser entero
df_ventas['cantidad'] = df_ventas['cantidad'].astype(int)
df_ventas
```

```python
# graficamos columna ventas de df_ventas
sns.lineplot(data=df_ventas, x='Fecha_Venta', y='cantidad')
# ajustamos el ancho de grafico a 12
plt.figure(figsize=(36, 6))
```

```python
# realizamos un histograma de cantidad
sns.histplot(data=df_ventas, x='cantidad')
```

```python
import numpy as np

# Definir la media y desviación estándar para concentrar los valores en 2 y 4
media = 3  # Punto medio entre 2 y 4
desviacion = 1  # Para que la mayoría de los valores caigan entre 2 y 4

# Generamos los valores con distribución normal
vendedores_normales = np.random.normal(loc=media, scale=desviacion, size=len(df_ventas))

# Redondeamos y limitamos los valores al rango [1, 5]
df_ventas['Vendedor'] = np.clip(np.round(vendedores_normales), 1, 5)

df_ventas
```

```python
sns.histplot(data=df_ventas, x='Vendedor')
```

```python
# Definir la media y desviación estándar para concentrar los valores en 2 y 4
media = 2  # Punto medio entre 1 y 4
desviacion = 1  # Para que la mayoría de los valores caigan entre 2 y 4

# Generamos los valores con distribución normal
productos_normales = np.random.normal(loc=media, scale=desviacion, size=len(df_ventas))

# Redondeamos y limitamos los valores al rango [1, 5]
df_ventas['Producto'] = np.clip(np.round(productos_normales), 1, 4)

df_ventas
```

```python
sns.histplot(data=df_ventas, x='Producto')
```

```python
# Generar valores aleatorios "Si" o "No" para la columna 'Envio'
df_ventas["Envio"] = np.random.choice(["Si", "No"], size=len(df_ventas))

# Mostrar las primeras filas
df_ventas
```

```python
sns.histplot(data=df_ventas, x='Envio')
```
